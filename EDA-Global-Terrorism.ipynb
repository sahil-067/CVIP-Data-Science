Taking only necessary columns for eda

```python
data=data[["iyear","imonth","iday","country_txt","region_txt","provstate","city",\
       "attacktype1_txt","targtype1_txt","targsubtype1_txt",\
       "gname","motive","weaptype1_txt","nkill","nwound"]]
```

Giving better names to columns

```python
data.rename(columns={"iyear":"Year","imonth":"Month","iday":"Day","country_txt":"Country",\
                   "region_txt":"Region","provstate":"Province_State","city":"City",\
                   "attacktype1_txt":"Attack Type","targtype1_txt":"Target Type",\
                   "targsubtype1_txt":"Target","gname":"Gang Name",\
                   "motive":"Motive","weaptype1_txt":"Weapon Type",\
                   "nkill":"Killed","nwound":"Wounded"},inplace=True)
```

again analysing the colunmns

```python
data.info()
```

Checking for null values in data

```python
data.isnull().sum()
```

Filling out NULL values

```python
data["Killed"]=data["Killed"].fillna(0)
data["Wounded"]=data["Wounded"].fillna(0)
data['Target Type'] = data['Target Type'].fillna('Unknown')
data['Target'] = data['Target'].fillna('Unknown')
data['Province_State'] = data['Province_State'].fillna('Unknown')
data['City'] = data['City'].fillna('Unknown')
data['Motive'] = data['Motive'].fillna('Unknown')
```

```python
data.isnull().sum()
```

## **Visualisation Starts Here**

Day wise people killed

```python
data[["Day","Killed"]].groupby("Day").sum().plot(kind="bar",color="green",figsize=(20,5))
plt.title("Day wise peoples killed")
plt.xlabel("Day")
plt.ylabel("Number of people Killed")
plt.show()
```

Month wise people killed

```python
data[["Month","Killed"]].groupby("Month").sum().plot(kind="bar",color="green",figsize=(20,5))
plt.title("Month wise peoples killed")
plt.xlabel("Month")
plt.ylabel("Number of people Killed")
plt.show()
```

number of people killed per year

```python
data[["Year","Killed"]].groupby("Year").sum().plot(kind="bar",color="green",figsize=(20,5))
plt.title("Year wise peoples killed")
plt.xlabel("Year")
plt.ylabel("Number of people Killed")
plt.show()
```

number of people got killed by each attack type

```python
data[["Attack Type","Killed"]].groupby("Attack Type").sum().sort_values(by="Killed").plot(kind="bar",color="green",figsize=(20,5))
plt.title("No of people got Killed in each Attack type")
plt.xlabel("Attack Types")
plt.ylabel("Number of people got killed")
plt.show()
```

number of people got killed by each weapon type

```python
data[["Weapon Type","Killed"]].groupby("Weapon Type").sum().sort_values(by="Killed").plot(kind="bar",color="green",figsize=(20,5))
plt.title("number of people got killed by each weapon type")
plt.xlabel("Weapon Types")
plt.ylabel("Number of people got killed")
plt.show()
```

number of people got killed for each Target type

```python
data[["Target Type","Killed"]].groupby("Target Type").sum().sort_values(by="Killed").plot(kind="bar",color="green",figsize=(20,5))
plt.title("number of people got killed for each target type")
plt.xlabel("Target Types")
plt.ylabel("Number of people got killed")
plt.show()
```

Number of people killed for all targets individually

```python
data[["Target","Killed"]].groupby("Target").sum().sort_values(by="Killed").plot(kind="bar",color="green",figsize=(20,5))
plt.title("Killed in each Region")
plt.xlabel("Target Name")
plt.ylabel("number of kills")
plt.show()
```

Number of people killed by each gang

```python
data["Gang Name"].value_counts()[0:18].plot(kind="bar",color="green",figsize=(20,5))
plt.title("Killed in each Region")
plt.xlabel("Region")
plt.ylabel("Number of people got killed")
plt.show()
```

Number of people killed in each Region

```python
data[["Region","Killed"]].groupby("Region").sum().sort_values(by="Killed").plot(kind="bar",color="green",figsize=(20,5))
plt.title("Killed in each Region")
plt.xlabel("Region")
plt.ylabel("Number of people got killed")
plt.show()
```

number of people killed in each Country

```python
data[["Country","Killed"]].groupby("Country").sum().sort_values(by="Killed").plot(kind="bar",color="green",figsize=(20,5))
plt.title("Killed in each Country")
plt.xlabel("Country")
plt.ylabel("Number of people got killed")
plt.show()
```

Terrorism in each region by year

```python
pd.crosstab(data.Year,data.Region).plot(kind='area',figsize=(20,5))
plt.title("terrorism in each region by year")
plt.xlabel("Year")
plt.ylabel("Number of people got killed")
plt.show()
```
