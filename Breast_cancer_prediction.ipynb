Importing important libraries and the sample dataset on breast cancer

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
```

Creating the DataFrame

```python
data=load_breast_cancer()
df = pd.DataFrame(data = data.data, columns = data.feature_names)
df["target"] = pd.Series(data.target)
```

Making Feature Matrix and Target matrix

```python
X=df.drop('target',axis=1)
Y=df['target']
```

Splittting data for training and testing

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

Perform the model tarining for LR model

```python
from sklearn.linear_model import LogisticRegression
lr_model=LogisticRegression()
lr_model.fit(X_train,Y_train)
```

Evaluating the LR model

```python
score=[]
sc=lr_model.score(X_test,Y_test)
print(sc)
score.append(sc)
```

```python
from sklearn.metrics import f1_score,classification_report
pred=lr_model.predict(X_test)

print(classification_report(Y_test,pred))
```

Going for Other models

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import ExtraTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier

mod={
    'Random Forest':RandomForestClassifier(),
    'Decision Tree':DecisionTreeClassifier(),
    'Extra Trees':ExtraTreeClassifier(),
    'SVC':SVC(),
    'Naive Bayes':GaussianNB(),
    'K-Nearest Neighbours':KNeighborsClassifier()
}
```

```python
for name,clas in mod.items():
  clas.fit(X_train,Y_train)
  sc=clas.score(X_test,Y_test)
  score.append(sc)
  print(f"{name} model has a score of {sc}")
```

```python
score
```

```python
a=["Logistic Regression",'Random Forest', 'Decision Tree', 'Extra Trees', 'SVC', 'Naive Bayes', 'K-Nearest Neighbours']
plt.figure(figsize=(12,6))
bar=plt.bar(a,score)

plt.xlabel('Model Names')
plt.ylabel('Score')
plt.title('Accuracy of different models')

plt.show()
```
